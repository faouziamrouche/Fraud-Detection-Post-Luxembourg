```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix
from sklearn.datasets import load_iris
from pandas import DataFrame
import numpy as np
import pandas
import pickle

event = pandas.read_csv('./Events.csv', sep=',')
events = event [:2000]
events
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
# define example

Time = events['Time']
SU = events['SU']
DU = events['DU']
SC = events['SC']
DC = events['DC']
AT = events['AT']
LT = events['LT']
AO = events['AO']
SF = events['SF']
onehot_encoded_global = Time
for i in SU, DU, SC, DC, AT, LT, AO, SF :
    values = array(i)
    #print(values)
    # integer encode
    label_encoder = LabelEncoder()
    integer_encoded = label_encoder.fit_transform(values)
    #print(integer_encoded)
    # binary encode
    onehot_encoder = OneHotEncoder(sparse=False)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    ### CONCATENATE
    onehot_encoded = pandas.DataFrame(onehot_encoded)
    onehot_encoded_global = pandas.concat([onehot_encoded_global, onehot_encoded], axis=1, sort=False)
    #print(onehot_encoded)
# invert first example
#inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
onehot_encoded_global = pandas.DataFrame(onehot_encoded_global)
print(len(onehot_encoded_global.columns))

```
